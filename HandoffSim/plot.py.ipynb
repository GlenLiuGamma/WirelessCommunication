```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
y1= np.loadtxt('best.txt', unpack=True)
y2= np.loadtxt('threshold.txt', unpack=True)
y3= np.loadtxt('entropy.txt', unpack=True)
x=np.arange(0,86401)
```

```python
line1, = plt.plot(x, y1, color = 'red', linewidth = 4, label = 'Best')             
line2, = plt.plot(x, y2, color = 'blue', linewidth = 2, label = 'Threshold')
line3, = plt.plot(x, y3, color = 'green', linewidth = 3, label = 'Entropy')
plt.legend(handles = [line1, line2, line3], loc='lower right')
plt.savefig("result.png",dpi=300,format="png")
```
